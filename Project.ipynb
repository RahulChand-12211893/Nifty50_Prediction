```python
#training an initial machine learning model
#using RandomForestClassifier for default version
#finding non lenear predication, like if taget is high then it is not necessary that price will rice


from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 2000, min_samples_split= 100, random_state = 1)


#n_estimator is number of individual decision tree, higher the value better the accuracy
#min_samples_split helps overfitting, highervthe value the less accurate it will be but more overfits
#random if model is rum more than twice then the ramdon number genetaed will be in predicted sequence we will get the same result is model is run twice.


#model is learning about the pediction 
#puting all of the rows except the last 100 rows into the training set
#puthe last 100 days into the test set
train = nifty50.iloc[:-100]
test = nifty50.iloc[:-100]
predictors = ["Close", "Volume", "Open","High","Low"]
model.fit(train[predictors], train["Target"])


#It will take some time
```

```python
#measuring how accurate our model is
#using precision sroce to generate predicting with the help of 'predict' method 

from sklearn.metrics import precision_score
preds = model.predict(test[predictors])


```

```python
#rather then last 100 days
#Building a backtesting system
#craeting a prediction function which will wrap up everyting we did into one function


def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combine = pd.concat([test["Target"], preds], axis = 1)
    return combine
```

```python
#writing backtesting function which takes in our nifty50 data, a macjine learning model,
# our predictors and start values and step value
# step = 250 indicate we have 250 training days and taking data to next 10 years thats why start = 2500
#to predict the data of 11 year

def backtest(data, model, predictors, start=2500, step=250 ):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)   
    
```

```python
#doing bacltest with the help of data and model and predictors


predictions = backtest(nifty50, model, predictors)
```

```python
# 1 denote market is up for how much days
# 0 denote market is down for how much days
#to get the value of exact 250 days, change return indexing



predictions["Predictions"].value_counts()
```

```python
#this will give our precision score
#we were 55% accurate


precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#as benchmark to check that 55% was good we are using this


predictions["Target"].value_counts() / predictions.shape[0]
```

```python
#adding additional predictors to our model to improve accuracy
#we will calculate the mean close price in the last two days, 5 days, 2 60 days
# then we will caluculate retion between tadays closing price in those periods which will
# help us to know, is the market gone up a ton because if so, it may be due for a downturn 


horizons = [2,5,60,250,1000]
new_predictors = []
for horizon in horizons:
    rolling_averages = nifty50.rolling(horizon).mean()
    
    ratio_column = f"Trend{horizon}"
    nifty50[ratio_column] = nifty50["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    nifty50[trend_column] = nifty50.shift(1).rolling(horizon).sum()["Target"]
    new_predictors += [ratio_column, trend_column]
```

```python

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    #[:,1] is the probability of stock price going up
    preds [preds >= .6] = 1
    preds[preds< .6] =0
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
predictions = backtest(nifty50, model, new_predictors)
```
